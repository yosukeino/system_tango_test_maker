```python
import pandas as pd

# CSVファイルの読み込み
df = pd.read_csv('list.csv', encoding="shift-jis")

# 基本情報の確認
print("=== データの基本情報 ===")
print(f"行数: {len(df)}")
print(f"列数: {len(df.columns)}")
print("\n=== 列名一覧 ===")
print(df.columns)

# データの中身を確認
print("\n=== データプレビュー ===")
display(df.head())
```

```python
# 不要な列を削除してから、indexを1から振り直す
cleaned_df = df.drop([col for col in df.columns if 'Unnamed' in col], axis=1)
cleaned_df.index = cleaned_df.index + 1  # indexを1ずつ増やして1から開始

# クリーニング後のデータを確認
print("=== クリーニング後のデータ ===")
print(f"行数: {len(cleaned_df)}")
print(f"列数: {len(cleaned_df.columns)}")
print("\n=== 列名一覧 ===")
print(cleaned_df.columns)
print("\n=== データプレビュー ===")
display(cleaned_df.head())
```

```python
import random

def create_custom_test(df, 
                      start_num=1,          # 開始番号
                      end_num=100,          # 終了番号
                      num_questions=50,      # 出題数
                      direction='ja_to_en',  # 'ja_to_en' または 'en_to_ja'
                      test_type='partial',   # 'partial'（一部空欄）または 'full'（全文）
                      random_select=True,    # 範囲内からランダム選択するか
                      random_order=False):   # 問題をランダムに並べ替えるか
    """
    カスタマイズされたテストを作成する関数
    
    Parameters:
    -----------
    df : DataFrame
        問題データを含むDataFrame
    start_num : int
        開始番号
    end_num : int
        終了番号
    num_questions : int
        出題数
    direction : str
        'ja_to_en': 日本語→英語
        'en_to_ja': 英語→日本語
    test_type : str
        'partial': 一部空欄
        'full': 全文記入
    random_select : bool
        範囲内からランダムに選択するか
    random_order : bool
        問題をランダムに並べ替えるか
    """
    # 指定された範囲のデータを取得
    available_indices = list(range(start_num, end_num + 1))
    
    # 問題の選択
    if random_select:
        # 範囲内からランダムに選択
        selected_indices = random.sample(available_indices, min(num_questions, len(available_indices)))
    else:
        # 範囲内から順番に選択
        selected_indices = available_indices[:num_questions]
    
    # 問題の並び替え
    if random_order:
        random.shuffle(selected_indices)
    else:
        selected_indices.sort()
    
    # テストのヘッダー表示
    print("=== カスタムテスト ===")
    print(f"範囲: {start_num}～{end_num}")
    print(f"問題数: {num_questions}")
    print(f"形式: {'日本語→英語' if direction == 'ja_to_en' else '英語→日本語'}")
    print(f"タイプ: {'一部空欄' if test_type == 'partial' else '全文'}")
    print("\n")
    
    # 問題の作成
    for question_num, idx in enumerate(selected_indices, 1):
        row = df.loc[idx]
        
        if direction == 'ja_to_en':
            # 日本語→英語の場合
            question = row['訳']
            if test_type == 'partial':
                answer_format = row['英作問題']
            else:
                answer_format = '________________________________'
        else:
            # 英語→日本語の場合
            question = row['フレーズ']
            if test_type == 'partial':
                answer_format = row['和訳問題']
            else:
                answer_format = '________________________________'
        
        print(f"{question_num}. {question}")
        print(f"   {answer_format}")
        print()

# 使用例
print("テスト例（デフォルト設定）:")
create_custom_test(cleaned_df)

print("\nテスト例（カスタム設定）:")
create_custom_test(cleaned_df,
                  start_num=1,
                  end_num=20,
                  num_questions=5,
                  direction='en_to_ja',
                  test_type='full',
                  random_select=True,
                  random_order=True)
```

```python
class TestMakerUI:
    def __init__(self, df):
        self.df = df
        self.max_index = len(df)
        self.setup_widgets()
        
    def setup_widgets(self):
        # 問題範囲選択（スライダー）
        self.range_slider = widgets.IntRangeSlider(
            value=[1, 20],
            min=1,
            max=self.max_index,
            step=1,
            description='問題範囲:',
            continuous_update=False,
            layout=widgets.Layout(width='400px')
        )

        # 出題数選択
        self.num_questions = widgets.IntSlider(
            value=20,
            min=1,
            max=50,
            step=1,
            description='出題数:',
            continuous_update=False,
            layout=widgets.Layout(width='400px')
        )

        # 問題形式選択
        self.direction = widgets.RadioButtons(
            options=[('日本語→英語', 'ja_to_en'), ('英語→日本語', 'en_to_ja')],
            value='ja_to_en',
            description='問題形式:',
            layout=widgets.Layout(margin='10px 0')
        )

        # テストタイプ選択
        self.test_type = widgets.RadioButtons(
            options=[('一部空欄', 'partial'), ('全文記入', 'full')],
            value='partial',
            description='解答形式:',
            layout=widgets.Layout(margin='10px 0')
        )

        # 問題の並べ替えのみ表示
        self.random_order = widgets.Checkbox(
            value=False,
            description='ランダムに問題を並べ替え',
            layout=widgets.Layout(margin='10px 0')
        )

        # テスト作成ボタン
        self.create_button = widgets.Button(
            description='テスト作成',
            button_style='primary',
            layout=widgets.Layout(width='200px', margin='20px 0')
        )
        
        # 結果表示エリア
        self.output = widgets.Output()
        
        # ボタンクリック時の動作を設定
        self.create_button.on_click(self.on_button_clicked)

    def on_button_clicked(self, b):
        """テスト作成ボタンがクリックされた時の処理"""
        with self.output:
            self.output.clear_output()
            start, end = self.range_slider.value
            
            create_custom_test(
                self.df,
                start_num=start,
                end_num=end,
                num_questions=self.num_questions.value,
                direction=self.direction.value,
                test_type=self.test_type.value,
                random_select=True,  # 常にTrue
                random_order=self.random_order.value
            )
    
    def display_ui(self):
        """UIを表示"""
        # ヘッダー
        header = widgets.HTML(value='<h2>テスト作成</h2>')
        
        # ウィジェットを縦に配置
        layout = widgets.VBox([
            header,
            self.range_slider,
            self.num_questions,
            widgets.HTML(value='<br>'),
            self.direction,
            widgets.HTML(value='<br>'),
            self.test_type,
            widgets.HTML(value='<br>'),
            self.random_order,
            self.create_button,
            self.output
        ])
        
        display(layout)

# UIの作成と表示
ui = TestMakerUI(cleaned_df)
ui.display_ui()
```

```python
def create_test_pdf(df, test_config, test_number=None):
    """
    PDF出力用のHTMLを生成する関数
    
    Parameters:
    -----------
    df : DataFrame
        問題データを含むDataFrame
    test_config : dict
        テスト設定（範囲、問題数、形式など）
    test_number : str
        テスト番号（未指定の場合は自動生成）
    """
    from datetime import datetime
    
    # テスト番号の生成（YYYYMMDD + 連番3桁）
    if test_number is None:
        date_str = datetime.now().strftime('%Y%m%d')
        test_number = f"{date_str}001"  # 連番は後で管理機能を追加
    
    html_template = f"""
    <!DOCTYPE html>
    <html>
    <head>
        <meta charset="UTF-8">
        <style>
            @page {{
                size: A4;
                margin: 2cm;
            }}
            body {{
                font-family: "Helvetica Neue", Arial, "Hiragino Kaku Gothic ProN", "Hiragino Sans", Meiryo, sans-serif;
                margin: 0;
                padding: 20px;
            }}
            .container {{
                background: white;
                padding: 20px;
            }}
            .header {{
                display: flex;
                justify-content: space-between;
                align-items: flex-start;
                margin-bottom: 30px;
                border-bottom: 2px solid #f0f0f0;
                padding-bottom: 20px;
            }}
            .title-section {{
                flex: 2;
            }}
            .test-title {{
                font-size: 24px;
                font-weight: bold;
                margin-bottom: 10px;
            }}
            .test-info {{
                font-size: 14px;
                color: #666;
            }}
            .test-number {{
                color: #888;
                font-size: 12px;
            }}
            .name-section {{
                flex: 1;
                text-align: right;
            }}
            .name-field {{
                border: 2px solid #e0e0e0;
                border-radius: 6px;
                padding: 12px 25px;
                min-width: 250px;
                min-height: 25px;
                position: relative;
                margin-top: 20px;
            }}
            .name-field:before {{
                content: '氏名';
                position: absolute;
                top: -10px;
                left: 10px;
                background: white;
                padding: 0 5px;
                color: #666;
                font-size: 14px;
            }}
            .problems-container {{
                display: flex;
                flex-direction: column;
                gap: 20px;
            }}
            .problem-row {{
                display: grid;
                grid-template-columns: 40px 0.8fr 1.2fr;
                gap: 20px;
                align-items: center;
            }}
            .number {{
                width: 32px;
                height: 32px;
                background: #f8f9fa;
                border-radius: 50%;
                display: flex;
                align-items: center;
                justify-content: center;
                font-weight: bold;
                color: #2c3e50;
            }}
            .question {{
                font-size: 14px;
                padding: 8px 15px;
                background: #f8f9fa;
                border-radius: 4px;
                line-height: 1.5;
            }}
            .answer-space {{
                border-bottom: 2px solid #e0e0e0;
                min-height: 35px;
                font-size: 16px;
                padding: 5px 0;
            }}
            @media print {{
                body {{
                    padding: 0;
                }}
                .container {{
                    padding: 0;
                }}
            }}
        </style>
    </head>
    <body>
        <div class="container">
            <div class="header">
                <div class="title-section">
                    <div class="test-title">英単語テスト</div>
                    <div class="test-info">出題範囲: {test_config['start_num']}～{test_config['end_num']}</div>
                    <div class="test-number">テスト番号: {test_number}</div>
                </div>
                <div class="name-section">
                    <div class="name-field"></div>
                </div>
            </div>
            <div class="problems-container">
    """
    
    # 問題の生成
    selected_indices = test_config['selected_indices']
    for i, idx in enumerate(selected_indices, 1):
        row = df.loc[idx]
        
        # 問題生成部分を以下のように修正
        # 問題文と解答欄の生成
        if test_config['direction'] == 'ja_to_en':
            question = row['訳']
            if test_config['test_type'] == 'partial':
                answer_format = row['英作問題'].replace('(', '(&nbsp;&nbsp;&nbsp;&nbsp&nbsp;&nbsp;&nbsp;&nbsp&nbsp;&nbsp;&nbsp;&nbsp&nbsp;&nbsp;&nbsp;&nbsp;').replace(')', '&nbsp;&nbsp;&nbsp;&nbsp&nbsp;&nbsp;&nbsp;&nbsp&nbsp;&nbsp;&nbsp;&nbsp&nbsp;&nbsp;&nbsp;&nbsp;)')
            else:
                answer_format = '___________________________'
        else:
            question = row['フレーズ']
            if test_config['test_type'] == 'partial':
                answer_format = row['和訳問題'].replace('(', '(&nbsp;&nbsp;&nbsp;&nbsp&nbsp;&nbsp;&nbsp;&nbsp&nbsp;&nbsp;&nbsp;&nbsp&nbsp;&nbsp;&nbsp;&nbsp;').replace(')', '&nbsp;&nbsp;&nbsp;&nbsp&nbsp;&nbsp;&nbsp;&nbsp&nbsp;&nbsp;&nbsp;&nbsp&nbsp;&nbsp;&nbsp;&nbsp;)')
            else:
                answer_format = '___________________________'
                
        html_template += f"""
                <div class="problem-row">
                    <div class="number">{i}</div>
                    <div class="question">{question}</div>
                    <div class="answer-space">{answer_format}</div>
                </div>
        """
    
    html_template += """
            </div>
        </div>
    </body>
    </html>
    """
    
    return html_template

# テスト設定の例
test_config = {
    'start_num': 1,
    'end_num': 20,
    'selected_indices': range(1, 21),  # 実際には選択された問題のインデックス
    'direction': 'ja_to_en',
    'test_type': 'partial'
}

# HTMLの生成
html_content = create_test_pdf(cleaned_df, test_config)

# HTMLファイルとして保存（確認用）
with open('test_preview5.html', 'w', encoding='utf-8') as f:
    f.write(html_content)
```
